#### Python if Statement

```python
# if condition:
#     if-block
```

The if statement checks the condition first.

If the condition evaluates to True, it executes the statements in the if-block. Otherwise, it ignores the statements.

```python
age = input('Enter your age:')
if int(age) >= 18:
    print("You're eligible to vote.")
```

```python
age = input('Enter your age:')
if int(age) >= 18:
    print("You're eligible to vote.")
    print("Let's go and vote.")
```

#### Python if…else statement

```python
# if condition:
#     if-block;
# else:
#     else-block;
```

```python
age = input('Enter your age:')
if int(age) >= 18:
    print("You're eligible to vote.")
else:
    print("You're not eligible to vote.")
```

#### Python if…elif…else statement

```python
# if if-condition:
#     if-block
# elif elif-condition1:
#     elif-block1
# elif elif-condition2:
#     elif-block2
# ...
# else:
#     else-block
```

```python
age = input('Enter your age:')

# convert the string to int
your_age = int(age)

# determine the ticket price
if your_age < 5:
    ticket_price = 5
elif your_age < 16:
    ticket_price = 10
else:
    ticket_price = 18

# show the ticket price
print(f"You'll pay ${ticket_price} for the ticket")
```

#### Introduction to Python Ternary Operator

```python
age = input('Enter your age:')

if int(age) >= 18:
    ticket_price = 20
else:
    ticket_price = 5

print(f"The ticket price is {ticket_price}")
```

```python
ticket_price = 20 if int(age) >= 18 else 5
```

The following syntax is called a ternary operator in Python:


```python
# value_if_true if condition else value_if_false
```

```python
# if condition:
#     value_if_true
# else:
#     value_if_true
```

Note that you have been programming languages such as C# or Java, and you’re familiar with the following ternary operator syntax:

condition ? value_if_true : value_if_false

```python
age = input('Enter your age:')

ticket_price = 20 if int(age) >= 18 else 5

print(f"The ticket price is {ticket_price}")
```

#### Using Python for loop to calculate the sum of a sequence

The following example uses the for loop statement to calculate the sum of numbers from 1 to 100:

```python
sum = 0
for num in range(101):
    sum += num

print(sum)
```

2. Second, the sum is added with the number from 1 to 100 in each iteration.

3. Finally, show the sum to the screen.

#### Python break

Sometimes, you want to terminate a for loop or a while loop prematurely regardless of the results of the conditional tests. In these cases, you can use the break statement:

Using Python break with for loop

```python
for index in range(0, 10):
    print(index)
    if index == 3:
        break
```

When you use the break statement in a nested loop, it’ll terminate the innermost loop. For example:

```python
for x in range(5):
    for y in range(5):
        # terminate the innermost loop
        if y > 1:
            break
        # show coordinates on the screen
        print(f"({x},{y})")
```

The following example uses the break statement inside a while loop.

It’ll prompt you for entering your favorite color. The program will stop once you enter quit:

```python
print('-- Help: type quit to exit --')
while True:
    color = input('Enter your favorite color:')
    if color.lower() == 'quit':
        break
```

#### Python continue

The continue statement is used inside a for loop or a while loop. The continue statement skips the current iteration and starts the next one.

The following shows how to use the continue statement in a for loop:

And the following illustrates how to use the continue statement in a while loop:

#### Using Python continue in a for loop example

```python
for index in range(10):
    if index % 2:
        continue

    print(index)
```

Note that the index % 2 returns 1 if the index is an odd number and 0 if the index is an even number.

#### Using Python continue in a while loop example

The following example shows how to use the continue statement to display odd numbers between 0 and 9 to the screen:

```python
counter = 0

while counter < 10:
    counter += 1
    
    if not counter % 2:
        continue
        
    print(counter)
```
